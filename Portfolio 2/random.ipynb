```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def print_inorder(node):
    if node:
        print_inorder(node.left)
        print(node.val, end=' -- '),
        print_inorder(node.right)

def print_preorder(node):
    if node:
        print(node.val, end=' -- '),
        print_preorder(node.left)
        print_preorder(node.right)

def print_postorder(node):
    if node:
        print_postorder(node.left)
        print_postorder(node.right)
        print(node.val, end=' -- '),

# Building the tree
a = Node('a')
b = Node('b')
c = Node('c')
d = Node('d')
e = Node('e')
f = Node('f')
g = Node('g')
h = Node('h')
i = Node('i')
j = Node('j')
k = Node('k')

# Setting up the tree structure
a.left = b
a.right = c
b.left = d
b.right = e
d.left = h
e.left = i
c.left = f
c.right = g
g.left = j
g.right = k

# Performing the traversals
print("Inorder traversal of the tree is:")
print_inorder(a)

print("\nPreorder traversal of the tree is:")
print_preorder(a)

print("\nPostorder traversal of the tree is:")
print_postorder(a)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Function to calculate Root Mean Squared Error (RMSE)
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Evaluate Model
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse_value = rmse(y_test, y_pred)
    return mse, rmse_value

# Evaluate Model Performance
mse_a, rmse_a = evaluate_model(model_a, X_test_a, y_test_a)  # Model A
mse_b, rmse_b = evaluate_model(model_b, X_test_b, y_test_b)  # Model B
mse_c, rmse_c = evaluate_model(model_c, X_test_c, y_test_c)  # Model C
mse_d, rmse_d = evaluate_model(model_d, X_test_d, y_test_d)  # Model D

# Print the results
print("Model A (Case 1 - Most correlated features):")
print("MSE:", mse_a)
print("RMSE:", rmse_a)
print("\nModel B (Case 1 - Least correlated features):")
print("MSE:", mse_b)
print("RMSE:", rmse_b)
print("\nModel C (Case 2 - Most correlated features):")
print("MSE:", mse_c)
print("RMSE:", rmse_c)
print("\nModel D (Case 2 - Least correlated features):")
print("MSE:", mse_d)
print("RMSE:", rmse_d)
```
